```python
from pyspark.ml.feature import StringIndexer, VectorAssembler
from pyspark.ml.clustering import KMeans
```

```python
indexer = StringIndexer(inputCol="value", outputCol="value_index")
indexed_df = indexer.fit(df).transform(df)
indexed_df.show()
```

```python
assembler = VectorAssembler(
    inputCols=["value_index"],
    outputCol="features"
)

final_df = assembler.transform(indexed_df)
final_df.show()
```

```python
kmeans = KMeans(
    k=2,
    seed=1,
    featuresCol="features"
)

model = kmeans.fit(final_df)
```

```python
predictions = model.transform(final_df)
predictions.select("value", "prediction").show()
```

```python
predictions.write.mode("overwrite").saveAsTable("kmeans_results")
```

```python
spark.table("kmeans_results").show()
```
